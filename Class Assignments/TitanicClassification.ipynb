```python
import pandas

dataset = pandas.read_csv('train.csv')

dataset
```

```python
dataset.info()
```

```python
columns_we_want = [
    dataset.columns[2],
    dataset.columns[4],
    dataset.columns[5],
    dataset.columns[7],
    dataset.columns[9],
    dataset.columns[11],
    dataset.columns[1]
]

reduced_dataset = dataset[columns_we_want]

reduced_dataset
```

```python
reduced_dataset.info()
```

```python
reduced_dataset = reduced_dataset.dropna(0, how = 'any')

reduced_dataset
```

```python
from sklearn.preprocessing import LabelEncoder

reduced_dataset[reduced_dataset.columns[1]] = LabelEncoder().fit_transform(reduced_dataset[reduced_dataset.columns[1]])
reduced_dataset[reduced_dataset.columns[5]] = LabelEncoder().fit_transform(reduced_dataset[reduced_dataset.columns[5]])

reduced_dataset
```

```python
from sklearn.model_selection import train_test_split
import warnings

warnings.filterwarnings('ignore')

input_features = reduced_dataset.drop('Survived', 1)
output_features = reduced_dataset['Survived']

input_features
```

```python
train_x, test_x, train_y, test_y = train_test_split(input_features, output_features, test_size = 0.3, random_state = 42)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

classifier = SVC()
classifier.fit(train_x, train_y)

predictions = classifier.predict(test_x)
print(accuracy_score(test_y, predictions))
```

```python
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier()
classifier.fit(train_x, train_y)

predictions = classifier.predict(test_x)
print(accuracy_score(test_y, predictions))
```

```python
from sklearn.naive_bayes import GaussianNB

classifier = GaussianNB()
classifier.fit(train_x, train_y)

predictions = classifier.predict(test_x)
print(accuracy_score(test_y, predictions))
```
